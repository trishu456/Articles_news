```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from textblob import TextBlob
import matplotlib.pyplot as plt

# Step 1: Load the dataset
def load_data(file_path):
    return pd.read_csv(file_path)

# Step 2: Text Preprocessing
def preprocess_text(df):
    # Combine 'title' and 'text' into a single column
    df['combined_text'] = df['title'] + ' ' + df['text']
    return df['combined_text']

# Step 3: Feature Extraction
def extract_features(texts):
    vectorizer = TfidfVectorizer(stop_words='english')
    tfidf_matrix = vectorizer.fit_transform(texts)
    return tfidf_matrix

# Step 4: Sentiment Analysis
def analyze_sentiment(texts):
    sentiments = [TextBlob(text).sentiment.polarity for text in texts]
    return sentiments

# Step 5: Popularity Scoring
def calculate_popularity(sentiments):
    return sum(sentiments) / len(sentiments) if sentiments else 0

# Main Execution
file_path = 'gossipcop_fake.csv'  # Replace with your dataset path
df = load_data(file_path)

# Ensure required columns are present
required_columns = ['id', 'news_url', 'title', 'tweet_ids', 'text']
if all(col in df.columns for col in required_columns):
    texts = preprocess_text(df)
    tfidf_matrix = extract_features(texts)
    sentiments = analyze_sentiment(texts)
    popularity_score = calculate_popularity(sentiments)

    print(f'Popularity Score: {popularity_score}')

    # Step 6: Visualization
    plt.figure(figsize=(10, 5))
    plt.plot(sentiments, marker='o', label='Sentiment Polarity')
    plt.title('Sentiment Analysis of News Articles')
    plt.xlabel('Article Index')
    plt.ylabel('Sentiment Polarity')
    plt.grid()
    plt.legend()
    plt.show()
else:
    print(f"Error: Dataset is missing required columns. Ensure {required_columns} are present.")
```

```python
import pandas as pd

# Load datasets from your file paths
real_news = pd.read_csv('gossipcop_real.csv')  # Replace with your file path
fake_news = pd.read_csv('gossipcop_fake.csv')  # Replace with your file path

# Display first few rows of each dataset
print("Real News Sample:")
print(real_news.head())
print("\nFake News Sample:")
print(fake_news.head())
```

```python
# Calculate the number of tweets for each article
real_news['num_tweets'] = real_news['tweet_ids'].apply(lambda x: len(str(x).split('\t')))
fake_news['num_tweets'] = fake_news['tweet_ids'].apply(lambda x: len(str(x).split('\t')))

# Add label column
real_news['label'] = 'real'
fake_news['label'] = 'fake'
```

```python
# Combine the datasets
combined_news = pd.concat([real_news[['id', 'num_tweets', 'label']],
                           fake_news[['id', 'num_tweets', 'label']]])

# Convert labels to binary for correlation analysis
combined_news['label_binary'] = combined_news['label'].map({'real': 1, 'fake': 0})

# Compute correlation
correlation = combined_news[['num_tweets', 'label_binary']].corr().iloc[0, 1]
print(f"Correlation between number of tweets and label (real/fake): {correlation}")
```

```python
# Top 5 most tweeted real news
top_real_news = real_news.sort_values(by='num_tweets', ascending=False).head(5)
print("\nTop 5 Real News Articles by Number of Tweets:")
print(top_real_news[['title', 'num_tweets']])

# Top 5 most tweeted fake news
top_fake_news = fake_news.sort_values(by='num_tweets', ascending=False).head(5)
print("\nTop 5 Fake News Articles by Number of Tweets:")
print(top_fake_news[['title', 'num_tweets']])
```

```python
import matplotlib.pyplot as plt

# Plot number of tweets for real and fake news
plt.figure(figsize=(10, 6))
plt.hist(real_news['num_tweets'], bins=30, alpha=0.5, label='Real News', color='blue')
plt.hist(fake_news['num_tweets'], bins=30, alpha=0.5, label='Fake News', color='red')
plt.title('Distribution of Number of Tweets for Real vs. Fake News')
plt.xlabel('Number of Tweets')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
